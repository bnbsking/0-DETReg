```python
import os, glob
import visualization as vis

print( sorted(os.listdir("/home/jovyan/nas-dataset/recycling")) )

def visInterval(imgFolder, antFolder, prefix, pre=6, nex=6):
    path     = "/home/jovyan/nas-dataset/recycling"
    imgPathL = sorted(glob.glob(f"{path}/{imgFolder}/*.jpg"))
    antPathL = sorted(glob.glob(f"{path}/{antFolder}/*.txt"))
    i = imgPathL.index(f"{path}/{imgFolder}/{prefix}.jpg")
    j = antPathL.index(f"{path}/{antFolder}/{prefix}.txt")
    assert i==j, (i,j)
    for k,(imgPath,antPath) in enumerate(zip(imgPathL[i-pre:i+nex+1], antPathL[i-pre:i+nex+1])):
        print(k)
        vis.show(imgPath, antPath)
```

```python
visInterval("DF_WuGu_Relay_36K/testset", "DF_WuGu_Relay_36K/testset", "20220616_100121_707", 0, 0)
```
